4. Explore a single categorical variable

```python
# Displays the variables of the dataframe
print("User Data Columns:")
print(user_data.columns.tolist())

print("\nOrder Data Columns:")
print(order_data.columns.tolist())
```

```python
# Let's explore the gender variable in users
# value_counts() gives the frequency distribution
gender_dist = user_data['gender'].value_counts()
print("Gender Distribution:")
print(gender_dist)
```

```python
# Make a bar chart for the frequency distribution
sns.countplot(data=user_data, x='gender')
plt.title('Frequency Distribution of Gender')
plt.show()
```

```python
# Frequency distribution of the 'age' variable
age_dist = user_data['age'].value_counts()
age_pct = user_data['age'].value_counts(normalize=True) * 100

print("Age Distribution (Counts):")
display(age_dist)

print("\nAge Distribution (Percentages):")
display(age_pct)

# Visualizing the age distribution
plt.figure(figsize=(10, 6))
sns.countplot(data=user_data, x='age', order=age_dist.index, palette='viridis', hue='age', legend=False)
plt.title('Frequency Distribution of Age')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.show()
```

```python
# Frequency distribution of the 'age' variable
age_dist = user_data['age'].value_counts()
print("Age Distribution:")
print(age_dist)

# Visualizing the age distribution
plt.figure(figsize=(10, 6))
sns.countplot(data=user_data, x='age', order=age_dist.index)
plt.title('Frequency Distribution of Age')
plt.show()
```

5. Explore a single interval variable

```python
# Find quantile or percentile for original_unit_price
quantiles = order_data['original_unit_price'].quantile([0.25, 0.5, 0.75, 0.9, 0.99])
print("Quantiles for Original Unit Price:")
print(quantiles)
```

```python
# Find the maximum price
max_price = order_data['original_unit_price'].max()
print(f"The maximum original unit price is: {max_price}")

# Find the records with the maximum price
# We use the comparison to slice the original dataset
max_price_records = order_data[order_data['original_unit_price'] == max_price]
display(max_price_records)
```

```python
# A histogram to explore the distribution of the interval variable
# original price

plt.figure(figsize=(10, 6))
sns.histplot(order_data['original_unit_price'], bins=50, kde=True)
plt.title('Distribution of Original Unit Price')
plt.xlabel('Original Unit Price')
plt.ylabel('Frequency')
plt.show()

# The results are strongly affected by the maximum value.
```

```python
# delete the maximum record
order_data = order_data[order_data['original_unit_price'] < max_price]
print(f"Maximum price record deleted. New maximum price: {order_data['original_unit_price'].max()}")
```

```python
# The histogram again
plt.figure(figsize=(10, 6))
sns.histplot(order_data['original_unit_price'], bins=50, kde=True)
plt.title('Distribution of Original Unit Price (Outlier Removed)')
plt.xlabel('Original Unit Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Delete the outliers using slicing
# Keeping only records where the original_unit_price is less than 500
order_data_filtered = order_data[order_data['original_unit_price'] < 500]

# Visualize the distribution of the filtered data
plt.figure(figsize=(10, 6))
sns.histplot(order_data_filtered['original_unit_price'], bins=50, kde=True)
plt.title('Distribution of Original Unit Price (Slicing < 500)')
plt.xlabel('Original Unit Price')
plt.ylabel('Frequency')
plt.show()
```

6. Explore the relationship between two interval variables

```python
# Make a scatterplot to explore the relationship between original and final price
plt.figure(figsize=(10, 6))
sns.scatterplot(data=order_data_filtered.sample(1000), x='original_unit_price', y='final_unit_price', alpha=0.5)
plt.title('Relationship between Original and Final Unit Price (Sample of 1000)')
plt.xlabel('Original Unit Price')
plt.ylabel('Final Unit Price')
plt.plot([0, 500], [0, 500], color='red', linestyle='--') # Equality line
plt.show()
```
